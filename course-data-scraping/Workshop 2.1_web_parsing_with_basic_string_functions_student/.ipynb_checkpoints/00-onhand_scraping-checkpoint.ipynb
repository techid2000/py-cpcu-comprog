```python
# ---- TO DO 1 : Code Here ----
url = 'https://www.chula.ac.th/en/academics/faculties-and-schools'
html = str(urllib.request.urlopen(url).read())

i = html.find("<!-- PAGE BODY -->")
e = html.find("<!-- END PAGE BODY -->")
cnt = 0
name = []
while(True):
    i = html.find("<strong>",i,e)
    if(i==-1): break
    j = html.find("</strong>",i,e)
    name.append(html[i+8:j])
    print(html[i+8:j])
    i = j; cnt += 1
print("Number of Faculties : ",cnt)













```

# Extra score (1%)

Print Current and Retired Faculty Members Name and Email 

from https://www.cp.eng.chula.ac.th/en/about/faculty/

*** Crawl only active members that have image ***
*** บุคลากรที่ไม่มี email ก็ให้เว้นว่างไว้ ***

ตัวอย่างผลลัพธ์ที่ได้ส่วนต้น

<img src="source/extra_1.png">

ตัวอย่างผลลัพธ์ที่ได้ส่วนท้าย

<img src="source/extra_2.png">

```python
print("----- Current Faculty Members -----")
url = "https://www.cp.eng.chula.ac.th/en/about/faculty/"
html = str(urllib.request.urlopen(url).read())

start = html.find("Current Faculty Members")
end = html.find("</table>",start)
cur = start
while(True):
    cur = html.find("<img",cur,end)
    if(cur==-1):
        break
    cur = html.find("<p>",cur)
    cur2 = html.find("<",cur+1)
    name = html[cur+3:cur2].strip(' ')
    cur3 = html.find("mailto:",cur)
    if(name[0]=='\\'): continue
    print(name)
    if(cur3!=-1):
        print(html[cur3:html.find("\"",cur3)].strip('mailto:'))
    cur = cur2
    
print("\n----- Retired Faculty Members -----")
start = html.find("Retired Faculty Members")
end = html.find("</table>",start)
cur = start
while(True):
    cur = html.find("<img",cur,end)
    if(cur==-1):
        break
    cur = html.find("<p>",cur)
    cur2 = html.find("<",cur+1)
    name = html[cur+3:cur2].strip(' ')
    cur3 = html.find("mailto:",cur)
    if(name[0]=='\\'): continue
    print(name)
    if(cur3!=-1):
        print(html[cur3:html.find("\"",cur3)].strip('mailto:'))
    cur = cur2
```
